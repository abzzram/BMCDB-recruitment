```python
#import prospective student list of faculty names from local folder. I concatenate it all into one long row of nanes
fac_list = pd.read_csv('./faculty_list (2).csv')
fac_list
```

```python
from collections import Counter
```

```python
#count faculty
pd.set_option('display.max_rows', 500)
df = fac_list
Counts = df.value_counts()
print(Counts)
#the names aren't consistent
```

```python
#this part is not the best way to do it (I'm new at python)
#first name a list of the from top 20 names appearing in Counts. I had to look in the original file that each of these names are actually a unique faculty. For example
#I chose to search for 'Henry' because if I do 'Ho' it would find other people that have 'ho' in their name.
Names = ('Albeck','Juliano','Torre', 'Knoepfler', 'Elva','Fisher','Shih','Kermit','Jawdat','Enoch','Barlow','Henry','Neil','Dennis','Gelli','Carlito','Olson','Janine','Andreas','Mark','Nicole')
Names
```

```python
#figure out how to count names
Na_ = 'Albeck'
df.Faculty.str.count(Na_).sum()
```

```python
#make list of counts  
Fac = []
for name in Names:
    Na_ = name
    C = df.Faculty.str.count(Na_).sum()
    #print(C)
    Fac.append(C)
```

```python
Faculty_list = {'Faculty':Names,'Count': Fac}
Faculty_list = pd.DataFrame(Faculty_list)
Faculty_list
```

```python
#sort the list
top_faculty = Faculty_list.sort_values('Count',ascending =0)
top_faculty
```
